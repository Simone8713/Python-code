#### Creating stacked bar charts for classification demographics


```python
# Group by department and income_group
dep_income_group = customer_profiling_merged.groupby(['department','income_group']).size().unstack()
```

```python
# sort by ascending
dep_income_group = dep_income_group.loc[dep_income_group.sum(axis = 1).sort_values(ascending = True).index]
```

```python
# check values
dep_income_group.head(25)
```

```python
# stacked bar chart 
dep_income_group = dep_income_group.plot.barh(stacked = True, title ='Department vs Income Groups')
```

```python
# Exporting bar chart 'dep_income_group'
dep_income_group.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'dep_income_group'))
```

```python
# creating a crosstab
crosstab_loyal_region = pd.crosstab(customer_profiling_merged['region_loc'], customer_profiling_merged['loyalty_flag'], dropna = False)
```

```python
# Sort the crosstab 
crosstab_loyal_region = crosstab_loyal_region.sort_values(by='Loyal customer', ascending=True)
```

```python
# Create a stacked barchart from the crosstab
bar_loyal_region = crosstab_loyal_region.plot(kind="barh", stacked=True, rot=0)
```

```python
# Exporting bar chart 'bar_profile_income_region'
bar_loyal_region.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyal_region'))
```
