##### 4. Create an exclusion flag for low-activity customers (customers with less than 5 orders) and exclude them from the data.

```python
#Creating a exclusion flag for 'max_order' columnn
df_instacard_merged.loc[df_instacard_merged['max_order'] >= 5, 'exclusion_flag'] = 'normal-activity customers'
```

```python
df_instacard_merged.loc[df_instacard_merged['max_order'] < 5, 'exclusion_flag'] = 'low-activity customers'
```

```python
#checking the results 
df_instacard_merged[['max_order', 'exclusion_flag']].head(60)
```

```python
df_instacard_merged[['max_order', 'exclusion_flag']].tail(120)
```

```python
# Creating a sample for low-activity customers
df_low_act_customers = df_instacard_merged[df_instacard_merged['exclusion_flag'] =='low-activity customers']
```

```python
# Creating a sample for normal-activity customers
df_nor_act_customers = df_instacard_merged[df_instacard_merged['exclusion_flag'] =='normal-activity customers']
```
