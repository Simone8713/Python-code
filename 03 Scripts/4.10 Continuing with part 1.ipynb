##### Creating scatterplots for the customer profiles combined with popular goods

```python
# Group by '_profile_age' and 'department'
customers_profile_age_department = customer_profiling_merged.groupby(['profile_age','department']).size().unstack()
```

```python
# Sort by ascendingg
customers_profile_age_department = customers_profile_age_department.loc[customers_profile_age_department.sum(axis = 1).sort_values(ascending = True).index]
```

```python
# check values:
customers_profile_age_department.head(10)
```

```python
# Steps for creating the bubbleplot:

# import extra library
import matplotlib.cm as cm

# Get 'departments' for plotting
departments = customers_profile_age_department.columns

# Get the index values as 'profile_age' for plotting
profiles = customers_profile_age_department.index

# Flatten the DataFrame values for plotting
x = customers_profile_age_department.values.flatten()

# Create a list of x-coordinates repeating each 'profile' for each 'department'
profile_indices = range(len(profiles))
x_coords = [i for i in profile_indices for _ in departments]

# Create a list of y-coordinates repeating each 'department' for each 'profile'
department_indices = range(len(departments))
y_coords = [i for _ in profile_indices for i in department_indices]

# Create bubble plot
plt.figure(figsize=(8, 10))
sc = plt.scatter(x_coords, y_coords, s=x/2000, c=x, alpha=0.5, cmap='viridis')  # 's' is for size of points
sc.set_cmap('plasma')
sc.set_clim(0, 1000000)
plt.xticks(profile_indices, profiles)  # Set x-ticks to 'profile_age'
# To improve readability, rotate the labels of the x axis and align the text horizontally to the left
plt.xticks(rotation=-25, ha="left")
plt.yticks(department_indices, departments)  # Set y-ticks to 'department'
plt.xlabel('Profile Age')
plt.ylabel('Consumer Goods')
plt.title('Bubble Plot of Profile Age vs Consumer Goods')
plt.grid(True)
plt.colorbar(label='Number of Customers')  # Add colorbar for point sizes
plt.tight_layout()
# Saving the plot as a png picture
plt.savefig(os.path.join(path, '04 Analysis','Visualizations', 'profile_age_cgoods.png')) 
plt.show()
```
