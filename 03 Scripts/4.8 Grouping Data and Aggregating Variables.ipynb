```python
#group a given dataframe by a given column
df.groupby('product_name')
```

```python
#Split the data into groups based on “department_id.” and apply the agg() function to each group to obtain the mean values for the “order_number” column.
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#different command using the mean, that gives the same results but in a different format
df.groupby('department_id')['order_number'].mean()
```

```python
#using the groupby with multiple aggregations:
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
#aggregating data with transform
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
#Avoiding limit on rows for the .head() function
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(21)
```

```python
#Creating a loyalty flag for 'max_order' columnn
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check values in your loc column 'loyalty_flag':
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#checking the results with head() on multiple columns:
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#Question 6: Create a spending flag for each user based on the average price across all their orders
```

```python
#First, use transform to create the mean price column:
ords_prods_merge['average_price_product'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
#View specific columns:
ords_prods_merge[['user_id', 'average_price_product']].head(50)
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price_product'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price_product'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#check values in loc column 'spending_flag':
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
#regular check
ords_prods_merge[['user_id', 'average_price_product', 'spending_flag']].head(100)
```

```python
#Question 7: Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column
```

```python
# Again,use transform to create a column, this time the median of “days_since_prior_order” 
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
#View specific columns:
ords_prods_merge[['user_id', 'median_days_between_orders']].head(50)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] > 10) & (ords_prods_merge['max_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#check values in loc column 'order_frequency_flag':
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
#regular check
ords_prods_merge[['user_id', 'median_days_between_orders', 'order_frequency_flag']].head(100)
```
